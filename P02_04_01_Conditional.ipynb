# Conditional, 조건문
'흐름을 제어한다'는 표현에서 '제어'란 '어떤 상황에서 어떻게 행동하라'라는 일련의 명령들을 말한다.  
프로그래밍에서 '어떤 상황인지를 파악한다는 것'은 그 상황에 부합하는 조건을 명료하게 제시해서 확인한다는 것이다.
## 불리언 함수
조건문(Conditional)이란 주어진 조건이 참(True)인지 거짓(False)인지에 따라 프로그램의 흐름을 제어하는 구문을 말한다.  
보통 참을 1, 거짓을 0으로도 쓰고 다음과 같이 이들에 대한 함수를 불리언 함수(Boolean Function)라 부른다.  
$\mathit{f:\{0, 1\}^{n} \to \{0, 1\}}$  
간단한 불리언 함수로는 많은 프로그래밍 언어들이 그러하듯 참과 거짓을 부정하는 부정(Negation)이고, 연산자는 느낌표(`!`)를 앞에 붙여서 참과 거짓을 반전시킨다.

```julia
println(true)
println(!true)
println(false)
println(!false)
```

합집합(`AND, &&`) 또는 교집합(`OR, ||`)을 이항 연산자(Binary Operator)로 사용한다.

```julia
println(true && true)
println(true && false)
println(false && true)
println(false && false)
println()
println(true || true)
println(false || true)
println(true || false)
println(false || false)
```

이변수 함수인 `AND`와 `OR`이 다변수 함수로 일반화된 것으로 `all()`과 `any()`함수가 있다.  
이름 그대로 "all"은 주어진 불리언의 벡터가 모두(All) 참일 때만 참을 반환하고, "any"는 주어진 불리언 벡터 중 하나라도(Any) 참이면 참을 반환한다.

```julia
println(all([true]))
println(all([true, true]))
println(all([true, true, true]))
println(all([true, true, false]))
println()
println(any([true]))
println(any([true, true]))
println(any([true, true, false]))
println(any([false, false, false]))
```

또, 한 가지 유용하고 자주 쓰이는 함수로 배타적 논리합(XOR: Exclusive OR) 함수가 있다.  
`xor()` 함수는 불리언값 중 참이 있을 때 참을 반환하되, 둘 다 참이면 거짓을 반환한다.  
`xor()` 함수는 이항 연산자($\veebar$)로도 사용할 수 있다.

```julia
println(xor(true, true))
println(xor(true, false))
println(xor(false, true))
println(xor(false, false))
println()
# ⊻ = \xor
println(true ⊻ true)
println(xor(true, false))
println(xor(false, true))
println(xor(false, false))
```

## if ~ else 문
조건문은 기본적으로 `if`와 `else` 키워드를 사용한다.  
`if` 키워드 뒤에 불리언값이 오고, `else`키워드 뒤에는 불리언값이 오지 않는다.  
즉, 일반적인 `if ~ else`문과 동일하다.  
단, 맨 마지막 `end`키워드 사용하여 닫아 주어야 한다는 점을 주의한다.  
`if`문 뒤에 괄호로 묶은 불리언을 사용해도 문제 없다.

```julia
if π > 3
    println("π is greater than 3")
else
    println("π is not greater than 3")
end

if (π > 3)
    println("π is greater than 3")
else
    println("π is not greater than 3")
end
```

## if ~ elseif ~ else
`elseif` 마찬가지로 다른 프로그래밍 언어들과의 차이 없이 사용 가능하며, `else` 없이 `elseif` 로 끝내고 `end` 를 사용하는 것 또한 가능하다.  
하나 이상의 조건을 확인하기 위해 `elseif` 키워드를 사용하였는데  
`if ~ elseif ~ else` 같은 조건문을 사용하다 보면 자연스럽게 `switch`키워드가 생각나는데, Julia에는 `switch`문이 없다.

```julia
e = exp(1)

if e > π
    println("e is greate than π")
elseif e > 3
    println("e is not greater than 3")
else
    println("e is not greater than π and 3")
end

# 위와 같이 else 조건에 문의 문장이 출력되어야 하지만 elseif 에서 조건문이 끝나기 때문에 문장이 출력되지 않는다.
if e > π
    println("e is greate than π")
elseif e > 3
    println("e is not greater than 3")
end
```

`if ~ else`처럼 사용할 수 있는 `ifelse()`함수가 있다.  
이 함수는 첫 번째 인자로 조건무을 받고, 조건이 참일 경우 두 번째 인자를 반환, 거짓 이라면 세 번째 인자를 반환한다.  
거의 같은 표현으로 삼항연산자(Ternary Operator)인(`? :`)도 제법 자주 쓰인다.

```julia
ifelse(e > π, e, π)
```

```julia
e > π ? e : π
```

## try ~ catch
**허락보다 용서가 쉽다**
`if ~ else`문이 조건에 따라 코드를 실행하는 것을 '허락'에 비유한다면, `try-catch`문은 일단 실행시켜 보고 에러가 나면 그때 수습하는 '용서'의 방식이라 볼 수 있다.  
  
다음 예시에 정의된 `force_add()` 함수는 두 벡터의 덧셈을 수행하되, 길이가 달라도 짧은 쪽을 무시하고 강제로 더하는 기능을 한다.  
본래의 덱터 덧셈은 길이가 다를 때를 허용하지 않지만, `try`문을 통해서 일단 각 자리별로 덧셈을 시도하고 길이가 짧은 쪽의 원소가 존재하지 않아서 덧셈이 안 된다면 `catch`문에서 해당 에러를 `err`로 받는다.  
그리고 그 에러가 `BoundsError`(배열의 범위를 벗어나서 참조를 시도해 발생한 에러)인 경우 벡터의 길이가 다르다고 메시지를 출력한 후 둘 중 아무 원소나 채워 넣는 방식으로 구현되어 있다.  

```julia
function force_add(x, y)
    n = max(length(x), length(y))
    z = zeros(n)

    for k = 1:n
        try
            z[k] = x[k] + y[k]
        catch err
            if isa(err, BoundsError)
                println("Diferent length vectors!")
            end

            try z[k] = x[k] catch end
            try z[k] = y[k] catch end
        end
    end

    return z
end
```

테스트 결과는 아래와 같다.

```julia
u = [1, 8, 0, 4, 2]
v = [3, 6, 9]

force_add(u, v)
```

한편 어떤 에러가 발생했든 반드시 실행하게끔 하는 `finally`키워드도 있다.  
이 키워드에 있는 코드는 해당 코드 블록에서 나가면서 그 이전의 `try ~ catch`문이 어떻게 작성되었는지와 무관하게 실행된다.  
예를 들어, 다음 코드는 Julia 세션 내부에서만 돌아가는 게 아니라 외부에 파일을 생성 및 수정한느 작업을 수행하고 `finally키키워드를 통해 앞에서 어떤 에러가 발생했든 반드시 파일을 닫아서 마무리하는 것을 보장한다.

```julia
file = open("target_file.txt", "w")

try
    println(file, "Hello world")
finally
    close(file)
end
```
