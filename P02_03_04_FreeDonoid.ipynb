# Free Monoid, 자유 모노이드
모노이드(Monoid)란 수학 전공자를 기준으로 해도 학부 3학년쯤이나 접하는 어려운 개념으로, 추상대수학(Abstract Algebra)에서 정의 정도를 스쳐 지나가듯 들어볼 수 있습니다.  
그런데 모노이드에 대해 검색을 해보면 오히려 컴퓨터과학과 관련된 문헌을 많이 찾을 수 있습니다.
## 문자열의 연산: *, ^
Julia에서 문자열의 병합(Merge)는 곱셈 연산(`*`)을 통해 이루어진다.  
가령 `"Hello"`와 `"word"`라는 두 문자열이 있다면 이들을 이어 붙이는 것은 간단히 `"Hello" * "world"` 로 수행할 수 있다.

````julia
"Hello" * "world"
````

거듭제곱(`^`) 연산을 통해 문자열 반복을 곱셈(`*`)처럼 사용할 수 있다.  
거듭제곱을 기준으로 왼쪽에 문자열, 오른쪽에 자연수를 두어 몇 번 반복할지 지시할 수 있다.

````julia
"Hello world!!" ^ 4
````

````julia
"Hwllo world!!" ^ 0
````

문자열의 병합을 문자열을 더한다(Add)라고 표현해도 의사소통에 문제가 없다.  
다른 프로그래밍 언어들은 더하기(`+`)를 통해 문자열을 병합하기도 하는데  
Julia는 생소한 곱셈(`*`)을 사용한다.  
이는 수학적으로 더 말이 되기 때문이다.
## 수학적 근거로서의 자유 모노이드
수학에서 수식을 다룰 때 𝑥와 𝑦가 곱해져 있다면 굳이 곱셈 기호를 남기지 않고 생략한다.  
$\mathit{x * y = xy}$  
여러 문자열들의 곱으로 나타난 것을 곱하면  
$\mathit{ab * xyz = abxyz}$  
이상할 것 없고 자연스럽게 보인다.  
$\mathit{hello * world = helloword}$  
단지 수식으로 무의미한 문자들의 나열이 아니라, 뜻을 가진 문자열이라고 라지라도 곱셈을 통해 이어붙기는 것에는 전혀 문제가 없다.  
오히려 수학적으로 말이 되는가를 기준으로 생각해 보면 문자열을 덧셈으로 이어 가는 게 오히려 이상할지도 모른다.  
  
Julia의 벡터가 벡터 공간에서 말하는 벡터인 것처럼 Julia의 문자열은 자유 모노이드(Free Monoid)라고 하는 대수적 구조의 원소로 볼 수 있다.  
벡터 공간보다 더 추상적으로 느길 수도 있지만, 한편으로는 대수학(代數學, Algebra)이라는 말 자체가 '수를 대신해서 글자를 쓰는 것에 관한 공부'인 만큼 이미 어느 정도 아는 개념이라고 생각하면 그다지 낯설 것도 없는 이야기 이다.  
예를 들어, 유한 집함 $mathit{A = \{a, b, ..., Y, Z\}}$를 알파벳(Alphabet)이라 할 때 이들의 곱으로 얻을 수 있는 모든 원소를 포함하는 모노이드를 자유 모노이드라 한다.  
그리고 이 자유 모노이드의 원소를 문자열(String)이라 하면 정확히 줄리아에서 다루는 문자열이다.
```
모노이드의 수학적 정의
공집합이 아닌 집한 𝑆의 원소들이 이항 연산 *에 대해 다음의 세 가지 규칙을 만족할 때, (𝑆, *)를 모노이드(Monoid)라 한다.

모든 𝑝, 𝑞, 𝑟 ∈ 𝑆에 대해서.
- 연산에 대한 닫힘: 𝑝와 𝑞가 𝑆의 원소면, 그 곱인 𝑝 * 𝑞 또한 𝑆의 원소다.
- 연산에 대한 결합법칙: (𝑝 * 𝑞) * 𝑟 = 𝑝 * (𝑞 * 𝑟)
- 대산에 산한 항등원: 𝑝 * "" = "" * 𝑝 = 𝑝를 만족하는 공백 ""이 존재한다.
```
보편적으로 추상대수에서 덧셈 기호를 쓰는 건 그 연산이 교환법칙을 따를 때, 즉 두 항의 순서가 상관없을 때 많이 쓰인다.  
문자열의 병합이란 앞과 뒤가 명확하고, 이에 덧셈 기호 대신 곱셈 기호를 사용하도록 한 것은 이론적으로 대단히 상식적인 결정이다.

````julia
p = "you"
q = "love"
r = "me"

println((p * q) * r)
println(p * (q * r))
println(p * "")
println("" * p)
````

## 문자열의 불변성
Julia의 문자열은 많은 객체지향형 언어들이 그러하듯 불변(Immutable)해서 바뀔 수 없다.  
바뀔 수 없다는 것은 이를테면 C나 Matlab처럼 문자열(String)을 문자(Character)의 배열로 보고 그중 하나를 인덱싱해서 수정하는 행위, 일르테면 다음과 같은 핸들링이 불가능하다는 것이다.  
다음과 같은 코드는 오류가 발생된다.

````julia
julian = "julian"
println(julian)

julian[end] = '.'
````

그렇다면 지금까지 사용해온 문자의 병합이나 `replace()`등의 문자열을 수정하는 함수는 어떻게 동작하는 것인가?  
두 문자열의 병합 p \* q는 사실 p 뒷부분에 q를 이어 붙여 '업데이트'한 문자열이 아니라 아예 pq = p * q라는 새로운 문자열을 반환한 것이다.  
이러한 pq의 존재성은 모노이드가 연산에 대해 닫혀 있으므로 자명하고, 문자열의 일부를 수정하는 함수들 역시 그때 그때 새로운 문자열을 찾아서 반환한다.  
자유 모노이드의 맥락으로 다시 돌아와 보자면, 새로운 문자열들은 사실 처음부터 무한 집합 속에 존재하고 있었던 것으로 볼 수 있다.  
이렇게 개념적으로 접근했을 때, 줄리아의 문자열이 불변성을 가지는 것은 아주 자연스러운 일이다.
