# Loops, 반복문
프로그래밍에서 조건문 만큼 중요한 것이 반복문 혹은 루프문(Loops)이다.  
반복문은 이름 그대로 특정 작업을 반복적으로 수행하는 역할을 하며, Julia에서는 `while`과 `for` 두 가지 방법으로 사용할 수 있다.
## While 문
`while`키워드는 '~하는 동안'이란 의미 그대로, 주어진 조건이 참일 때 계속해서 반복한다.  
다음 예는 어떤 수 k를 0부터 1까지 증가시키면서 출력하는데, k가 10이상으로 커져서 조건이 거짓이 되면 반복문을 멈추게 된다.

```julia
k = 0

while k < 10
    k += 1
    println(k)
end
```

반복문 그 자체는 아니지만 반복문의 내부에서 반복문의 제어를 도와주는 키워드로 '탈출하다'라는 의미의 `break` 키워드를 사용해 반복문에서 탈출 가능하다.

```julia
k = 0

while k < 10
    k += 1

    if mod(k, 3) == 0
        break
    end

    println(k)
end
```

k가 3의 배수면 `break`키워드를 통해 반복문이 중단된다.  
실제로 1과 2만 출력 후 중단된 결과가 확인된다.  
  
탈출의 반대로 '계속하다'라는 의미의 `continue` 키워드를 만나면 키워드 이후에 코드는 실행 되지 않고 즉시 반복문의 시작 위치로 돌아간다.  
반복 중 처음 위치로 돌아가는 걸 '계속하다'라고 말하는 이유는 해당 반복이 종료되지 않았으며, 끝까지 반복된 것도 아닌 상태로 끝나지 않고 계속 이어진다는 의미에서 `break`와 대비되는 표현이다.

```julia
k = 0

while k < 10
    k += 1

    if mod(k, 3) == 0
        continue
    end

    println(k)
end
```

k가 3의 배수가 되면 continue 키워드를 만나서 반복문의 시작으로 돌아갔기 때문에 그 하위에 존재하는 코드를 실행하지 않아, 3의 배수가 출력되지 않은 것을 확인할 수 있다.
## For 문
`for`키워드 역시 '~에 대해'라는 뜻 그대로, 주어진 집합의 원소들에 대해 반복 잡업을 수행한다.  
`while`문과 다른 점이 있다면 근본적으로 반복이 '무한하지 않다'라는 점이다.  
`while`문은 반복이 언제 끝날지 확신할 수 없을 때, `for`문은 구체적이고 유한한 작업을 수행할 때 쓰인다.

```julia
for n = 1:10
    println(n)
end
```

Julia의 `for`문에서 한 가지 조금 특별한 점을 꼽는다면, 반복으로 사용하는 변수를 할당할 때 등호(`=`) 대신 `in`, 수학 집합론에서 원소의 포함을 나타낼 때 쓰이는 `∈`('\in' 입력 후 \<tab\>) 기호를 사용해서 `for`문을 실행할 수 있다는 점이다.  
`in`은 Python을 비롯한 많은 언어들도 채택하고 있지만, `∈`를 반복문의 공식적인 문법으로 포함시킨 부분은 Julia답다라는 생각이 드는 대목이다.

```julia
for n in 0:9
    print(n)
end

println()

for n ∈ 9:-1:0
    print(n)
end
```

`for`문은 기본적으로 '증가'하도록 동작하는데 위와 같이 '감소'하도록 동작하게 하려면  
조건을 위와같이 `시작값:조건(증가/감소):종료값`형태로 사용해야 한다.  
좀더 예를 보면 다음과 같다.

```julia
for n in 0:2:9
    print(n)
end

println()

for n in 9:4:0
    print(n)
end

println()

for n in 9:-4:0
    print(n)
end

println()

for n in 9:-10:0
    print(n)
end
```

## Iterator, 반복자
반복자란 말 그대로 반복에 관련된 개념이다.  
Python을 비롯한 객체지향 언어들만큼 깊게 공부할 필요는 없고, 자주 유용하게 쓰여서 꼭 알고있어야 할 함수 두 개만 울고 넘어가도록 한다.  
  
`enumerate()`함수는 '열거하다'라는 의미로 `for`문에서 주어진 배열뿐만 아니라 해당 배열의 인덱스도 함께 사용할 수 있도록 해준다.  
다음의 예시에서 보면 소수의 배열을 받아서 몇 번째 소수인지 또한 변수로 받아서 사용하고 있다.

```julia
prime = [2, 3, 5, 7, 11, 13, 17, 19]

for (k, p) in enumerate(prime)
    println("$k: $p")
end
```

`zip()`함수는 길이가 같은 여러 배열을 묶어주는 역할을 한다.  
다음의 예시에서는 소수의 배열과 피보나치수열을 묶어서 각자를 별개의 변수로 받아서 쓰고 있다.

```julia
fibnc = [1, 1, 2, 3, 5, 8, 13, 21]

for (f, p) in zip(prime, fibnc)
    println("$f - $p")
end
```

그 외에, Julia에서 each-가 붙은 함수들은 많은 경우 이러한 반복자들을 제공한다.  
꼭 빌트인 함수가 아니라 패키지에서도 이러한 관습을 따르는 편이니, 어떤 반복을 자주사한다는 생각이 들면 우선 관련된 함수를 찾아본느 습관을 들이는게 좋다.  
  
많은 함수 중 간단한 예로 `eachrow()`와 `eachcol()`함수는 행렬의 각 행, 열 그대로 반복자를 쓸 수 있게 해준다.  
물론 행렬의 크기를 확인해서 구체적으로 인덱스를 주고 한 행 한 행 참조해도 되지만, 복잡한 계산이 많아질수록 이러한 고수준 프로그래밍을 통해 코드의 가독성을 높이고 실수를 줄여줄 필요가 있다.

```julia
M = [ 4 8 9
      7 1 3
      2 5 6 ]

for mat in eachrow(M)
    println(mat)
end

println()

for rix in eachcol(M)
    println(rix)
end
```
